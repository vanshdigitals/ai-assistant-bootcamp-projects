AI Assistent ChatBot Development

```python
import ollama
```

```python
response = ollama.chat(
    model='mistral',
    messages=[
        {
            'role': 'user',
            'content': 'what is deep learning?'
        }
    ]
)
print(response['message']['content'])
```

```python

user_input = input("Ask question ... ")
response = ollama.chat(
    model='mistral',
    messages=[
        {
            'role': 'user',
            'content': user_input
        }
    ]
)
print(response['message']['content'])

```

```python
def gym_trainer(question):
    response = ollama.chat(
        model='mistral',
        messages=[
            {
                'role': 'system',
                'content': """You are a gym trainer.

Rules:
- Be polite and friendly
- Keep answers short and to the point
- Give a diet plan for weight loss"""
            },
            {
                'role': 'user',
                'content': question
            }
        ]
    )
    return response['message']['content']
```

```python
print(gym_trainer("i want a diet plan to lose weight"))
```
